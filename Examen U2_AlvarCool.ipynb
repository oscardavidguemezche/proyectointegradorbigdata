```python
## Relacionar datos

relacion = df[['artists','danceability','energy','explicit']]
relacion
```

```python
## Identificar datos

### Mediana
mediana = df["danceability"].median()

### Promedio
media = df["danceability"].mean()

### Varianza
varianza = df["danceability"].var(ddof=0)

### Minimo
minimo = df["danceability"].min()

### Maximo
maximo = df["danceability"].max()
```

```python
## Relacion lineal positiva

correlacion = df.corr(method='spearman')
correlacion
```

```python
sns.heatmap(correlacion, xticklabels = correlacion.columns, yticklabels = correlacion.columns, cmap='coolwarm')
```

```python
### Como podemos observar en el grafico, hay ciertos patrones de relacion entre las columnas por ejemplo:
 ### La columna energia esta bastante relacionado con el ruido, lo cual nos indica que entre mas ruidosa sea la cancion mas energica es.
 ### Caso contrario con energia y acustica, en donde podemos observar un claro contraste ya que no presenta mucha relacion
```
